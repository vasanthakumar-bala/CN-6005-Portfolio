```python
import pandas as pd

# Dataset
data = {
    'EmploymentStatus': ['Employed','Unemployed','Employed','Employed','Unemployed'],
    'CreditHistory': ['Good','Bad','Good','Bad','Good'],
    'IncomeLevel': ['High','Low','Medium','Medium','Low'],
    'LoanApproved': ['Yes','No','Yes','No','Yes']
}

df = pd.DataFrame(data)

# Calculate prior probabilities
prior = df['LoanApproved'].value_counts(normalize=True)

# Conditional probability function
def conditional_prob(feature, value, target):
    return len(df[(df[feature] == value) & (df['LoanApproved'] == target)]) / len(df[df['LoanApproved'] == target])

# New input
features = {
    'EmploymentStatus': 'Employed',
    'CreditHistory': 'Good',
    'IncomeLevel': 'Medium'
}

yes_prob = prior['Yes']
no_prob = prior['No']

for f, v in features.items():
    yes_prob *= conditional_prob(f, v, 'Yes')
    no_prob *= conditional_prob(f, v, 'No')

print("Prediction:", "Yes" if yes_prob > no_prob else "No")
```
